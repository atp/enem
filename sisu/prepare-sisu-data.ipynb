# Ler os arquivos do Sisu

```python
# ler xlsx é muito devagar, vamos transformar em csv
# rodar esta célula somente uma vez no seu repo. Demora....

import os
def xls2csv(fn):
    if fn.split('.')[-1] != 'xlsx':
        return
    if fn.split('_')[0] == 'orig/2010':
        df = pd.read_excel(fn, skiprows = 4)
        for ano in range(2010,2019):
            for edicao in [1,2]:
                vers = f"{ano}/{edicao}"
                dfc = df[df['EDIÇÃO'] == vers]
                fn = f"orig/{ano}_{edicao}_sisu.csv"
                dfc.to_csv(fn,index=False)
    else:
        df = pd.read_excel(fn, sheet_name=1)
        parts = fn.split('.')
        fn = ''.join(parts[:-1]) + '.csv'
        df.to_csv(fn,index=False)
    return

#xls2csv('orig/2010_2018_sisu.xlsx')
```

```python
for fn in os.listdir('orig/'):
    #print(fn)
    xls2csv(os.path.join('orig',fn))
```

## Função para buscar os dados de 2010 à 2024

```python
def notasSisu(ano,edicao):
    arquivo = f"orig/{ano}_{edicao}_sisu.csv"
    df = pd.read_csv(arquivo)
    if ano >= 2019:
        colunas = ['NO_IES','SG_IES','DS_CATEGORIA_ADM','SG_UF_CAMPUS','NO_CURSO','DS_TURNO','QT_VAGAS_CONCORRENCIA','QT_INSCRICAO','NU_NOTACORTE']
        if ano >= 2024:
            colunas = ['NO_IES','SG_IES','DS_CATEGORIA_ADM','SG_UF_CAMPUS','NO_CURSO','DS_TURNO','QT_VAGAS_OFERTADAS','QT_INSCRICAO','NU_NOTACORTE']
        novocols = ['ies','sg_ies','catadm','uf','curso','turno','vagas','inscricoes','notacorte']
        # TODO: generalizar para filtrar por cotas
        df = df[df['DS_MOD_CONCORRENCIA'] == 'Ampla concorrência']
        df = df[colunas]
        df.columns = novocols
    else:    
        colunas = ['NOME IES','SIGLA IES','NOME CURSO','TURNO','NOTA DE CORTE']
        novocols = ['ies','sg_ies','curso','turno','notacorte']
        df = df[df['TIPO MODALIDADE'] == 'Ampla Concorrência']
        df = df[colunas]
        df.columns = novocols
    return df

notasSisu(2024,1)
```

```python
def buscaCurso(curso,ano,edicao,regiao=None):
    curso = curso.upper()
    df = notasSisu(ano,edicao)
    df = df[(df['curso'].str.contains(curso,regex=False))]
    if regiao:
        regiao = regiao.upper()
        df = df[(df['uf'] == regiao) & (df['notacorte'] > 0)]
    return df
    
```
