```python
import pandas as pd

def read_enem_data():
    dataframes = {}  

    for i in range(2009, 2023):
        file_path = f"data/enem_1_{i}.csv"  # Caminho do arquivo CSV
        try:
            df = pd.read_csv(file_path)  # LÃª o arquivo CSV
            dataframes[f"enem_{i}"] = df  # Adiciona ao dicionÃ¡rio com a chave 'enem_XXXX'
        except FileNotFoundError:
            print(f"Arquivo {file_path} nÃ£o encontrado.")

    return dataframes

# Chame a funÃ§Ã£o para ler os DataFrames
enem_dataframes = read_enem_data()

# Agora vocÃª pode acessar os DataFrames individualmente, por exemplo:
# df_2009 = enem_dataframes["enem_2009"]
# df_2010 = enem_dataframes["enem_2010"]

# Ou iterar sobre eles:
for year, df in enem_dataframes.items():
    print(f"Ano: {year}")
    print(df.head())  # Exibe as primeiras linhas do DataFrame
```
