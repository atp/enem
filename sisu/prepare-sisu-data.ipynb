# Ler os arquivos do Sisu

## Função para buscar os dados de 2010 à 2024

```python
def notasSisu(ano,edicao):
    if ano >=2019:
        ano = str(ano)
        edicao = str(edicao)
        arquivo = "orig/" + ano + "_" + edicao + "_sisu.xlsx"
        colunas = ['NO_IES','SG_IES','DS_CATEGORIA_ADM','SG_UF_CAMPUS','NO_CURSO','NU_NOTACORTE','DS_TURNO']
        df = pd.read_excel(arquivo, sheet_name=1)
        df = df[df['DS_MOD_CONCORRENCIA'] == 'Ampla concorrência']
        df = df[colunas]
    else:
        ano = str(ano)
        edicao = str(edicao)
        vers = ano + "/" + edicao
        arquivo = "orig/2010_2018_sisu.xlsx"
        colunas = ['NOME IES','SIGLA IES','NOME CURSO','NOTA DE CORTE','TURNO']
        df = pd.read_excel(arquivo, skiprows = 4)
        df = df[df['EDIÇÃO'] == vers]
        df = df[colunas]
    return df
```

```python
notasSisu(2013,2)
```

## Função para busca de cursos

```python
def buscaCurso(curso,regiao,ano,edicao):
    curso = curso.upper()
    df = notasSisu(ano,edicao)
    df = df[(df['NO_CURSO']==curso)]
    if regiao != '':
        regiao = regiao.upper()
        df = df[(df['SG_UF_CAMPUS'] == regiao) & (df['NU_NOTACORTE'] > 0)]
    return df
    
```

```python
buscaCurso(curso='arquitetura e urbanismo',regiao='sp',ano=2019,edicao=1)
```
