```python
def read_enem_data():
    dataframes = {}

    for ano in range(2009, 2023):
        file_path = f"data/enem_1_{ano}.csv"
        df = pd.read_csv(file_path)
        dataframes[f"enem_{ano}"] = df
    return dataframes

enem_dataframes = read_enem_data()

column_name = "NU_NOTA_MT"

# PreparaÃ§Ã£o dos dados para o grÃ¡fico
years = []
mean_scores = []

for year, df in enem_dataframes.items():
    years.append(year)
    mean_score = df[column_name].mean()  # Calcula a mÃ©dia da coluna "NU_NOTA_MT"
    mean_scores.append(mean_score)

# CriaÃ§Ã£o do grÃ¡fico de pontos
plt.figure(figsize=(10, 6))
plt.scatter(years, mean_scores, c='blue', marker='o', alpha=0.7)
plt.xlabel("Ano")
plt.ylabel("MÃ©dia das Notas de MatemÃ¡tica")
plt.title("ComparaÃ§Ã£o da MÃ©dia das Notas de MatemÃ¡tica ao longo dos anos")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Exibe o grÃ¡fico
plt.show()
```
