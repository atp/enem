```python
import pandas as pd
import seaborn as sns
```

```python
notas_selecao = pd.read_csv('data/enem19-sem_Zzero.csv')
notas_selecao
```

### Encontrei que a menor nota de corte de medicina em 2022 foi 773,83 na ufersa. 
#### A mesma tem pesos de redação = 3, CN = 7, CH = 4, LC = 4, MT = 2.

## Cálculo da média 

```python
notas_selecao['media'] = (7*notas_selecao['NU_NOTA_REDACAO']+4*notas_selecao['NU_NOTA_CH']
                           +2*notas_selecao['NU_NOTA_MT']+4*notas_selecao['NU_NOTA_LC']+3*notas_selecao['NU_NOTA_REDACAO'])/20
```

```python
notas_selecao.head()
```

```python
#Retorna apenas os alunos que tiraram mais que 773.83
alunos_selecionados = notas_selecao[notas_selecao.media>=773.83]
alunos_selecionados
```

```python
alunos_selecionados.describe()
```

```python
len(alunos_selecionados)
```

## Porcentagem de alunos que passariam em medicina na UNIFASA

```python
porcentagem = round((len(alunos_selecionados)/len(notas_selecao))*100,2)
porcentagem
```

### Percebemos que de acordo com nossa amostra apenas 2.85 por cento dos estudantes seriam capazes de passar na faculdade federal com a menos nota no sisu (não estamos considerando avgas com cotas).
